```python
Items={1:{"Product":"Biscuit","Quantity":10,"Cost":10},
                 2:{"Product":"Detergent","Quantity":20,"Cost":75},
                 3:{"Product":"Rice","Quantity":10,"Cost":50},
                 4:{"Product":"Ice cream","Quantity":100,"Cost":20},
                 5:{"Product":"Candy packet","Quantity":10,"Cost":50}}

while True:
    print("::::::::::::::::::::ITEAMS AVAILABLE IN THE STORE::::::::::::::::::::\n")
    print ("%10s %12s %8s %8s"%("S.no","  Product  ","Quantity","Cost"))
    for i in Items:
        print("%10s %12s %8s %8s"%(i,Items[i]["Product"],Items[i]["Quantity"],Items[i]["Cost"]))
    A=int(input("How many packet of biscuit required"))
    Items[1]["Quantity"]-=A
    if Items[1]["Quantity"]<0:
        print("stock out")
    B=int(input("How many packet of Detergent required"))
    Items[2]["Quantity"]-=B
    if Items[2]["Quantity"]<0:
        print("stock out")
    C=int(input("How many packet of Rice required"))
    Items[3]["Quantity"]-=C
    if Items[1]["Quantity"]<0:
        print("stock out")
    D=int(input("How many packet of Ice cream required"))
    Items[4]["Quantity"]-=D
    if Items[4]["Quantity"]<0:
        print("stock out")
    E=int(input("How many packet of Candy packet required"))
    Items[5]["Quantity"]-=E
    if Items[5]["Quantity"]<0:
        print("stock out")
        break
    print()
    print("\nEnter your Delivery details")
    Name=input("Please enter your name")
    Address=input("Please enter your address")
    distance=int(input("Enter distance from Your home 5/10/15"))
    if distance<5:
        delivery_charge=0
    elif distance>=5 and distance<10:
        delivery_charge=20
    else:
        delivery_charge=40
    
    print("\nDelivery Charge : {0} will be levied for {1} km of distance".format(delivery_charge,distance))
    Item_order={1:{"Product":"Biscuit","Quantity":A,"Total":A*10},2:{"Product":"Detergent","Quantity":B,"Total":B*75},3:{"Product":"Rice","Quantity":C,"Total":C*50},4:{"Product":"Ice cream","Quantity":D,"Total":D*20},5:{"Product":"Candy packet","Quantity":E,"Total":E*50}}  
    print ("%10s %12s %8s %8s"%("S.no","  Product  ","Quantity","Total"))
    Total_amount=0
    for i in Item_order:
        print ("%10s %12s %8s %8s"%(i,Item_order[i]["Product"],Item_order[i]["Quantity"],Item_order[i]["Total"]))
        Total_amount+=Item_order[i]["Total"]
    print("Total_amount----",Total_amount) 
    print("Final Bill including delivery charge----",Total_amount+delivery_charge)
    print("\nname----",Name)
    print("\nAddress",Address)
    Decision=str(input("Do you want to continue yes/no"))
    if Decision=="yes":
        continue
    else:
        break
print("\n :::::::::::::::::Thank You!",Name,":::::::::::::::::::::::::")

    
    
    
    
```
